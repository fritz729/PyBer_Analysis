```python
# Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import statistics 


#Load in csv

ride_data_df = pd.read_csv("Resources/PyBer_ride_data.csv")
ride_data_df


```

```python
# setting ticks location 
x_axis = np.arange(len(ride_data_df))
tick_location = [value for value in x_axis]

# plot data
ride_data_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_location, ride_data_df["Month"])
plt.show()

```

```python


# Same data but plot in bar graph 
ride_data_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', color="blue")
plt.show()

#spacing ticks
ride_data_df.arange(0, 50, step=5.0)



# #standard deviation of y values 
# stdev = statistics.stdev(y)
# stdev
# #adding error bars
# plt.bar(x, y, yerr=stdev, capsize=3)
# plt.show()
```
